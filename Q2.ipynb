#### Import library and files

```python
import pandas as pd

inventory_data = pd.read_csv('inventory_data.csv')
majors_data = pd.read_csv('majors_data.csv')
occupancy_data = pd.read_csv('occupancy_data.csv')
persons_data = pd.read_csv('persons_data.csv')
```

#### Handling duplicate majors

```python
majors_data
```

```python
majors_data = majors_data.drop_duplicates(subset=['name'])
```

```python
majors_data
```

#### Handling Name

```python
persons_data['name'] = persons_data['firstName'] + ' ' + persons_data['lastName']
```

```python
persons_data.drop(columns=['firstName', 'lastName'], inplace=True)
```

#### Handling Address

```python
persons_data[['address1', 'city_state']] = persons_data['address'].str.split(',', 1, expand=True)
```

```python
persons_data['address2']=''
```

```python
persons_data[['city', 'state']] = persons_data['city_state'].str.split(',', 1, expand=True)
```

```python
persons_data.drop(columns=['address', 'city_state'], inplace=True)
```

```python
persons_data['zip']=''
```

```python
persons_data
```

#### Drop duplicate PersonIDs

```python
persons_data = persons_data.drop_duplicates(subset=['personId'])
```

```python
persons_data
```

#### Majors to majorIds

```python
majors_name_to_id = pd.Series(majors_data.id.values, index=majors_data.name).to_dict()
def map_majors_to_ids(majors_str):
    if pd.isna(majors_str):
        return ''
    majors_list = [majors_name_to_id.get(maj.strip(), '') for maj in majors_str.split(',')]
    return ','.join(filter(None, majors_list))
```

```python
persons_data['majorIds'] = persons_data['majors'].apply(map_majors_to_ids)
```

```python
persons_data
```

```python
persons_data.drop(columns=['majors'], inplace=True)
```

```python
persons_data
```

#### Including bedIds

```python
occupancy_with_bedId = pd.merge(occupancy_data, inventory_data, on=['buildingName', 'roomName', 'bedName'], how='left')
```

```python
occupancy_with_bedId
```

```python
final_data = pd.merge(persons_data, occupancy_with_bedId[['personId', 'bedId']], on='personId', how='left')
```

```python
final_data
```

#### Cleaning data based on null or duplicate majorIds and bedIds and finally removing duplicate emails

```python
cleaned_data = final_data
cleaned_data = cleaned_data[['personId', 'name', 'email', 'dob', 'address1', 'address2', 'city', 'state', 'zip', 'majorIds','bedId']]
cleaned_data = cleaned_data.dropna(subset=['personId', 'name', 'email', 'dob', 'majorIds', 'bedId'])
cleaned_data = cleaned_data[cleaned_data['personId'] != '']
cleaned_data = cleaned_data[cleaned_data['name'] != '']
cleaned_data = cleaned_data[cleaned_data['email'] != '']
cleaned_data = cleaned_data[cleaned_data['dob'] != '']
cleaned_data = cleaned_data[cleaned_data['majorIds'] != '']
cleaned_data = cleaned_data[cleaned_data['bedId'] != '']
cleaned_data
```

```python
cleaned_data = cleaned_data.drop_duplicates(subset=['email'])
cleaned_data
```

#### Excluded rows

```python
excluded_rows = final_data.merge(cleaned_data, on=list(final_data.columns), how='left', indicator=True)
excluded_rows = excluded_rows[excluded_rows['_merge'] == 'left_only'].drop(columns=['_merge'])
```

```python
excluded_rows
```

```python
cleaned_data.to_csv('cleaned_data.csv', index=False)
```

```python
excluded_rows.to_csv('excluded_rows.csv', index=False)
```
